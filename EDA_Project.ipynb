# **Project Summary -**

The dataset under analysis consists of 125,354 records and 19 columns, representing detailed information about movies and TV shows along with associated metadata such as ratings, runtime, genres, and cast details. This dataset combines content-level and person-level information, making it suitable for comprehensive exploratory data analysis (EDA).

The initial inspection of the dataset reveals a mix of categorical and numerical features. Categorical variables include attributes such as title, type (movie or TV show), genres, and production countries, while numerical variables include release year, runtime, IMDb scores, TMDB scores, and popularity metrics. This diversity enables both statistical analysis and visualization.

A crucial part of the analysis involved identifying missing values. Several columns contain significant null values, notably age_certification (over 68,000 missing entries) and seasons (over 116,000 missing entries). The high number of missing values in the seasons column indicates that a large portion of the dataset consists of movies rather than TV shows. Other fields such as IMDb scores, TMDB scores, and character names also contain missing values, which require appropriate handling through imputation or removal depending on the context.

The dataset also contains 168 duplicate records, which can affect analysis accuracy if not removed. Data cleaning steps such as removing duplicates and handling null values are essential to ensure the reliability of further analysis.

Exploratory visualization plays a key role in understanding the dataset. Histograms of numerical variables such as release year, runtime, IMDb scores, and TMDB popularity provide insights into data distribution. For instance, the release year histogram shows that most content is concentrated in recent decades, indicating the growth of digital streaming platforms. The runtime distribution highlights variability in content length, reflecting the presence of both short-form and long-form media.

The distribution of IMDb scores reveals that most content falls within a moderate rating range, suggesting an overall average quality of content with fewer extreme ratings. Similarly, TMDB popularity is highly skewed, where a small number of titles have exceptionally high popularity while most remain moderately viewed.

Additional visualizations such as bar charts for content type show the proportion of movies versus TV shows, confirming that movies dominate the dataset. Genre-based analysis highlights the most frequently occurring genres, which can provide insights into audience preferences and industry trends. Time-series analysis of release years further emphasizes the increasing production of content over time.

Overall, the dataset demonstrates characteristics typical of real-world data: incompleteness, redundancy, and variability. Through systematic EDA, these issues are identified and addressed, enabling a clearer understanding of the data structure and underlying patterns.

In conclusion, this analysis transforms raw, unstructured data into meaningful insights by applying data cleaning, statistical summaries, and visualizations. The findings reveal trends in content production, rating distributions, and popularity, providing a strong foundation for further analysis, such as recommendation systems or predictive modeling.
